```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('hotel_bookings 2.csv')
df
```

```python
df.info()
```

```python
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'],format='mixed')
```

```python
df.describe(include=object)
```

```python
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
df.isnull().sum()
```

```python
df.drop(['company','agent'],axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df=df[df['adr']<5000]
```

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts())
```

```python
plt.figure(figsize=(8,5))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
plt.title('Resevation Status in Different Hotels')
plt.xlabel('Hotel')
plt.ylabel(' Numbers of Reservation')
plt.legend(['not_canceled','canceled'])
plt.show()
```

Want to know in a percentage form


```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Avg Daily Rate In City & Resort Hotel')
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='city Hotel')
plt.legend()
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette=['yellow','brown'])
```

```python
plt.figure(figsize=(16,8))
plt.title('ADR per month')

# Create the barplot with custom colors
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), 
            palette=['grey', 'purple'])

plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(10,7))
plt.title('Top 10 country reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.legend()
plt.show()
```

```python
canceled_df = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df.reset_index(inplace=True)
canceled_df.sort_values('reservation_status_date', inplace=True)

not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_df = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df.reset_index(inplace=True)
not_canceled_df.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(12, 5))
plt.title('Avg Daily Rate')
plt.plot(canceled_df['reservation_status_date'], canceled_df['adr'], label='Canceled', color='red')
plt.plot(not_canceled_df['reservation_status_date'], not_canceled_df['adr'], label='not_Canceled', color='green')
plt.legend()
plt.show()
```
